### Data Basics
* The formal name for a row is a case or observational unit. 
* The columns represent characteristics, called variables

* Variable Classification




* categorical variable but the levels have a natural ordering.called an ordinal variable, 

* while a regular categorical variable without special ordering is called a nominal variable.

* Relationships between variables
  + When two variables show some connection with one another, they are called associated variables. Associated variables can also be called dependent variables
  + A pair of variables are either related in some way (associated) or not (independent). No pair of variables is both associated and independent
  + Positive Association-> one increases other increases. For negative vice versa

* Explanatory and response variables
 + When we suspect one variable might causally affect another, we label the first variable the explanatory variable and the second the response variable.
 
* Randomized Experiment
  + Book Defination is terrible
  + A random experiment is a process that generates well-defined experimental outcomes.On any single repetition or trial, the outcome that occurs is determined completely by chance.
  + E.g. Toss Coin.  Well-defined outcomes : Head or Tails. For One Toss, Occurrence of H/T is decided by probability/chance.
  + The sample space for a random experiment is the set of all experimental outcomes e.g. { Heads, Tails }
  

### Sampling Principles and Strategies
* Population
* Sample
* Anecdotal Evidence
  + Data that is not clearly representative of population e.g Death of person due to heart attack, when doctor gave him new meds
    so med does not work.
* Sampling from Polulation
  + For Representative Sample of polulation, sample elements should be chosen randomly.
  + If there is preference, then it has *Bais* e.g. We connect to people within our reach
  + simple random sample : Each Case in population has equal chance of being included, It helps to minimize bias
  + With Simple Random Sample, Polution may not be accuratly represented by sample. Leads to High *non-response rate*
  + convenience sample are easily accesible cases.
 
* Sampling Methods
* All stastatical methods have notion of implied randomness.
* Four Methods of Sampling 
  1. Simple Random Sampling : Each Case has same chance of getting included.
  2. Stratified Sampling
     + Divide and Conqure kind of sampling . Population into Groups/stratas
     + Each Strata are chosen by grouping similar cases 
     + For choosing Starta use simple random sampling
     + Only choose when cases of each stratus are similar wrt outcome of interest 
  3. Cluster Sample 
     + Break Population into many cluster/groups
     + Sample fix number of clusters and include all observations from each of clusters into sample
  4. Multistage sample 
     + Same as cluster, but while creating sample instead of all observation, chose random sample within each selected cluster.

* Modelling Variability Problems
+ For Given 2 Variables from any Experiment, we need to claim relationship between them statistically
+ (Step 1)Two Claims
    1. Two variables are independent of each other : Independence model.(H0) 
    2. Two variables are *not* independent Alternative model(HA)

+ (Step 2)Conclusion
    1. Study result does not provide strong result against H0
    2. Study results evidence sufficiently strong to reject H0
    
   
+ In Formal study, Notion of observing rare event is rejected, so we reject independence model  in favour of alternative. Other way to say is "data provides strong evidence"

+ (Step 3) Conclude differences evaluated are by chance
