## Measuring Economy

### Gross Domestic Products
*  *money value* of all  *final goods and services* produced by factors of production located within an economy *over given period of time*
* Non Formal / Non Market Transactions are not measured .eg. services from  housewife, unpaid services e.g. linux 
* Illegal transactions like black market , smuggliing etc are not measured

* Financial Transactions like
  + Securities stocks and bonds
  + Govt transfers for scholarships and compensation 
  + Money Gifts
  
* Scale with wrong calibration, used consitently will provide useful results
* Benefit
  + Per Capita GDP -> Standard of Living
  + availabilty of Goods
  + Identification fo sectoral and regional growth
  
### 3 Ways of Measuring GDP 
1.Expenditure Approach
* National Income =  Consumer + Investment + Govt Expenditure  + Export - Import  ( C+I+G+X - M)

2.Income Approach
* National Income =  Rent + Interest + Wages + Profit

3.Production Approach
* National Income = sum of all values added at each stage of production by deducting value of intermediate  products from output
* Wheat rs 50 -> bread rs.70 -> sandwitch rs. 100  , NI = 50 + 20 +30 = 100.

### Measures of National Productivity 
* Capital Consumption Allowance or Depreciation  (Gross or Net)
* People within national boundary or citizen (Domestic or Natinal)
* Indirect Taxes - Subsides (Market or Factor)
