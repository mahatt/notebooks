### 3 Ways of Measuring GDP 
1.Expenditure Approach
* National Income =  Consumer + Investment + Govt Expenditure  + Export - Import  ( C+I+G+X - M)

2.Income Approach
* National Income =  Rent + Interest + Wages + Profit

3.Production Approach
* National Income = sum of all values added at each stage of production by deducting value of intermediate  products from output
* Wheat rs 50 -> bread rs.70 -> sandwitch rs. 100  , NI = 50 + 20 +30 = 100.

* *Nominal price* - Absolute price of a good, unadjusted for inflation
* *Real Price*  - Price of a good adjusted against the inflation indices
* E.g. real, or inflation-adjusted price of chips went up, they mean that the price of chips went up by more than overall inflation.
* Inflation means that the nominal prices of all goods and services in the economy increase on average
* Nominal GDP = current year output with current year prices
* Real GDP = current year output with base year prices


* Consumer Price Index (CPI) : Measure of the aggregate price level at consumers’ level
* (CPI) is the cost of a standard basket of goods and services consumed by a typical domestic consumer.
* it is weighted sum of prices 
* to determine the importance of different  expenditures by the average household for Industrial Worker, agricultural worker

* Producer Price Index (WPI) : Measure of the aggregate price level for intermediate products and wholesale goods
    + wholesale price index (WPI) is a price index of goods and services domestically produced.
    + measures the average level of prices of goods sold by producers.
    + Measures from perspective of seller
    

* Inflation between year Y1 and Y2 =  Price in Y2 / Price in Y1
* Nomial Price / Real Price at year Y1 =  CPI at Year Y1 / CPI at Base Year
