```python
""" Arithmatic Mean Age of person attending course """
YrOfExpDataframe =  all_table['Total Exp (In Years)'].head(31)
print(YrOfExpDataframe)
print(YrOfExpDataframe.sum() /  (YrOfExpDataframe.size-1))
print("calculating mean")
print(YrOfExpDataframe.mean())
```
