# Distributions

## random experiment
* a process that generates well-defined experimental outcomes. On any single repetition or trial, the outcome that occurs is determined completely by chance.
*  Tossing Coin is random exp but Throwing stone is not

## Probability Distribution
* The probability distribution for a random variable describes how probabilities aredistributed over the values of the random variable. 
* For a discrete random variable x, a probability function  f (x), provides the probability for each value of the random variable 

*  For Rolling a fair dice, Sample Space size is 6 with x belonging to Sample Space PDF f(x) is 1/6

* Properties :  sum of f(x) = 1 and f(x) >= 0
* Discrete Uniform Probability Function  f(x) =  1 / Sample Size

* Expected Value E(x) , Mean of Discrete Random Variable , is central tendency of Variable
 
* Bivariate Distributions :  Has  two random variables eg Rolling two dices
* Co-variance of Two random variables =   (variance(x+y) -variance(x) -variance(y) ) * 1/2
* Correlation bof two random variables  =  (variance(x +y)) / (variance(x)* variance(y)  takes values between [1,-1], from Strongly Correlated to Strongly not Correlated, 0 indicates no correlation


## Discrete Probabilty Distribution

### Binomial Distribution
*  Binomial Experiment 
    + Has sequence of n indentical trials
    + For Each trial, Only two outcomes are possible
    + Probabilty of Success / failure does not change from trail to trial
    + Trials are independent
* Number of experimental Outcomes providing exactly "x" suceess in n trials is nCx = n!/(x! * (n-x)!)

* Probability of x success in n trials is Binomial PDF = f(x) = nCx * p^x * (1-p)^(n-x) ; p is prob of success in one trial, 

### Poisson Probability Distribution
* Poisson PD 
    +  models random arrivals in waiting line 
    + Number of occurences over interval of time and space
    + Probability of an occurence is same for any two intervals of equal length 
    + Occurence or Non-occurence in any interval is independent of any other interval
    
* Number of arrivals in car wash or Number of leaks in 100 miles  of pipeline


## Continous Probability Distribution
* For random variable  
  + in Discrete, probabilty function f(x) provides prob that random variable assumes particular value
  + In Continuous, *Probabilty density* functions does not provide prob, but area under curve provides prob value in interval
  
### Uniform Distribution
* PDF f(x)
  +  f(x) = 1/(b-a)  for a<= x <=b
  +  0 otherwise

* Area as Measure of Probability = Height  * Width 

### Normal Distribution
* Unimodal ,symmetric and bell curve
* ubiquitous
* Parameters  Mean and Standard Deviation
* PDF

* Mean = median = mode!
* Normal Dist has Skewness zero

* Increase std dev -> wider,flat curve
* Area under curve =  1

* 1 Std dev = 68.3%
* 2 std dev = 95.4%
* 3 Std Dev = 99.7%

* std normal dist is mean = 0 and std dev = 1

* Tree types of Query for probabilites
  + Probabilty that std normal random variable z  <= given value
  + Probabilty that std normal random variable  v1 < z < v2
  + Probabilty that std normal random variable z >= given value
  


```python
import numpy as np
import statistics
import matplotlib.pyplot as plt
from  scipy.stats import norm

x =  np.arange(-2,2,0.01)
mean =  statistics.mean(x)
stdev = statistics.stdev(x)
plt.plot(x,norm.pdf(x,mean,stdev))

x =  np.arange(-1,1,0.01)
mean =  statistics.mean(x)
stdev = statistics.stdev(x)
plt.plot(x,norm.pdf(x,mean,stdev))

x =  np.arange(-3,3,0.01)
mean =  statistics.mean(x)
stdev = statistics.stdev(x)
plt.plot(x,norm.pdf(x,mean,stdev))

plt.show()
```
