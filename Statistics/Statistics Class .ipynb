```python
import pandas as pd
all_table = pd.read_csv(r'C:\Users\Maheshri\OneDrive\Documents\PGCBM.csv')
all_table.head(31)
```

```python
""" Arithmatic Mean Age of person attending course """
YrOfExpDataframe =  all_table['Total Exp (In Years)'].head(31)
print(YrOfExpDataframe)
print(YrOfExpDataframe.sum() /  (YrOfExpDataframe.size-1))
print("calculating mean")
print(YrOfExpDataframe.mean())
```

```python
""" Median Age of Person attending course"""
print("Panda Median :")
print(YrOfExpDataframe.median())
import math
sortedExp = YrOfExpDataframe.sort_values()
print(sortedExp)
print(sortedExp[math.floor(sortedExp.size / 2 )])
```
