Objectives
* Marketing Strategy

### Segmentation


* Effective Segmentation Requirements
    + Measurable  size, purchasing power and profile
    + Accessible for reach and serve  
    + Substantial  large or profitable enough
    + Differentiable 
    + Actionable effective programs can be designed to attract and serve segment
  

### Targetting

* Evaluate segment size and growth, segment structural attractiveness, and company objectives and resources
* Select Segment
    + A target market consists of a set of buyers who share common needs or characteristics that a company decides to serve
    

* CHoosing Targetting Strategy 
    + Limited resources  ->  Concentrated
    + the degree of product variability, Uniform products -> undifferentiated 
    + market variability
    + competitors marketing strategies, if competitor is differentiated dont use undifferntiated!
    
### Positioning
* product position is the way a product is defined by consumers on important attributesthe place the product occupies in consumers minds relative to competing products.
* perceptual positioning maps that show consumer perceptions of their brands versus those of competing products on important buying dimensions.


#### Choosing Positioning Strategy
* Identifying Possible Value Differences and Competitive Advantages
*  Which Differences to Promote
    + Important : Highly valued to target consumer
    + Distinctive 
    + Superior 
    + Communicable
    + Preemptive
    + Affordable
    + Profitable
    
 
* value propositionthe full mix of benefits on which a brand is differentiated and positioned
* It is the answer to the customers question Why should I buy your brand?


#### Developing a Positioning Statement
* To (target segment and need) our (brand) is ( concept) that (point of difference)
* e.g. Evernote is positioned as a digital content management application that helps busy people to capture and remember moments and ideas and find them fast later.
