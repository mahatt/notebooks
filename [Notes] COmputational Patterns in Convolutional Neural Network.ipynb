## Discrete Convolution
  + Does not process topological feature 
  + it is linear transformation (affine) that preserves ordering
  + Sparse : only a few input units contribute to a given output unit
  + reuses parameters :the same weights are applied to multiple locations in the input
  + Input Feature Map x Kernel =  Output Feature Map   For single feature
  + Light Blue : Input
  + Dark Blue : Kernel
  + Green : Output 
  
  
  
  
  + For Multiple Features  Repeat Above process and bind all output feature maps.
  + Dimentionality is Number of Input Feature Maps
  + For each of Input Feature Map, kernel can be different
  + Resulting Feature map is elementwise sum

## Transpose Convolution Arithmatic or Deconvolution
* Transformation in opposite direction of Convolution
* e.g. Convolutional encoder as input to decoder !
*  fractionally strided convolutions
#### Express  Convolution as Matrix Operation

* Sparse Matrix C =  wi,j of the kernel (with i and j being the row and column of the kernel respectively)


* kernel decides convolution, Based of FWD or BKW pass, Conv or Decolv is decided
* possible to emulate a transposed convolution with a direct convolution
* transposed convolution on a given input is to imagine such an input as being the result of a direct convolution applied on some initial feature map.
* allows to recover the shape 3 of this initial feature map
* One way to understand the logic behind zero padding is to consider the connectivity pattern of the transposed convolution and use it to guide the design of the equivalent convolution. For example, the top left pixel of the input of the irect convolution only contribute to the top left pixel of the output, the top  right pixel is only connected to the top right output pixel, and so on. 
* To maintain the same connectivity pattern in the equivalent convolution it is necessary to zero pad the input in such a way that the first (top-left) application  of the kernel only touches the top-left pixel, i.e., the padding has to be equal to the size of the kernel minus one.

## Dilation Convoluation
* inflate the kernel by inserting spaces between the kernel elements
*  used to cheaply increase the receptive field of output units without increasing the kernel size, which is especially effective when multiple dilated convolutions are stacked one after another.
