#### The Implications for Managers
* POWER
* To minimize incentive system facilitated misconduct, managers should do their best to hire employees who have appropriately tempered preference structures
* To minimize culture-facilitated misconduct, managers should do their best to hire employees who have appropriate values
* To minimize formal power facilitated misconduct in organizations

* STRUCTURE
* To minimize administrative system facilitated misconduct, managers should be alert to the possible unintended consequences of the controls that they put in place. 
* To minimize accidental wrongdoing,managers should, of course, do their best to insure that work environments arewell designed and organizational participants well-equipped to navigate these environments, such that human error, miscommunication, and faulty coordination are held to a minimum.
